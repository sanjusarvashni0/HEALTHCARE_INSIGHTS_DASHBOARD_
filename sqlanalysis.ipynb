```python
import sqlalchemy as db


engine = db.create_engine('mysql+mysqlconnector://root:root@localhost:3306/healthcare_data')
```

```python
import pandas as pd
data = pd.read_csv('C:/sanju/Healthcare_Insights_Dashboard/Healthcare-datasets-updated.csv')
data.head()
```

```python
data.to_sql(name="patients_data", con=engine, if_exists="replace", index=False)
```

```python
query  = "SELECT * from patients_data"
```

```python
df = pd.read_sql(query, con=engine)
```

```python
df
```

```python
# Trends in admission over time

query = "SELECT DATE_FORMAT(Admit_Date, '%Y-%m') AS Month, COUNT(*) AS Total_Admissions FROM patients_data GROUP BY Month ORDER BY Month;"
df = pd.read_sql(query, con=engine)

df
```

```python
#Diagnosis Frequency Analysis: Identify the top 5 most common diagnoses.

query_2 = "SELECT Diagnosis, COUNT(*) AS Diagnosis_Count FROM patients_data GROUP BY Diagnosis ORDER BY Diagnosis_Count DESC LIMIT 5;"
df2 = pd.read_sql(query_2, con=engine)
df2
```

```python
#Bed Occupancy Analysis: Analyze the distribution of bed occupancy types.

query_3 = "SELECT Bed_Occupancy, COUNT(*) AS Total_Admissions FROM patients_data GROUP BY Bed_Occupancy ORDER BY Total_Admissions DESC;"
df3 = pd.read_sql(query_3, con=engine)
df3
```

```python
# Length of Stay Distribution: Analyze the average and maximum length of stay for patients.

query_4 = "SELECT AVG(DATEDIFF(Discharge_Date, Admit_Date)) AS Average_Stay, MAX(DATEDIFF(Discharge_Date, Admit_Date)) AS Max_Stay FROM patients_data;"
df4 = pd.read_sql(query_4, con=engine)
df4
```

```python
#Seasonal Admission Patterns: Identify the seasonality in admissions based on the month.

query_5 = "SELECT MONTH(Admit_Date) AS Month, COUNT(*) AS Total_Admissions FROM patients_data GROUP BY Month ORDER BY Total_Admissions DESC;"
df5 = pd.read_sql(query_5, con=engine)
df5
```

```python
#Doctor Workload Analysis
query_6 = "SELECT Doctor,  COUNT(*) AS Total_Patients FROM patients_data GROUP BY Doctor ORDER BY Total_Patients DESC;"
df6 = pd.read_sql(query_6, con=engine)
df6
```

```python
#Revenue Analysis - an overview of hospital revenue and how much patients are paying out-of-pocket versus insurance-covered amounts.
query_7 = "SELECT SUM(`Billing Amount`) AS Total_Revenue, SUM(`Health Insurance Amount`) AS Total_Insurance_Covered, SUM(`Billing Amount` - `Health Insurance Amount`) AS Out_Of_Pocket_Payments FROM patients_data;"
df7 = pd.read_sql(query_7, con=engine)
df7
```

```python
# Most Expensive Diagnoses - Identifies which diagnoses have the highest treatment costs, aiding in budgeting and patient financial assistance planning
query_8 = "SELECT Diagnosis,  AVG(`Billing Amount`) AS Avg_Billing_Amount FROM patients_data GROUP BY Diagnosis ORDER BY Avg_Billing_Amount DESC LIMIT 5;"
df8 = pd.read_sql(query_8, con=engine)
df8
```

```python
# Follow-up Rate by Diagnosis - Shows the percentage of patients who require follow-ups for each diagnosis, useful for understanding post-discharge care needs.
query_9 = "SELECT Diagnosis, COUNT(`Followup Date`) AS Total_Followups, COUNT(*) AS Total_Cases,(COUNT(`Followup Date`) / COUNT(*)) * 100 AS Followup_Rate FROM patients_data GROUP BY Diagnosis ORDER BY Followup_Rate DESC;"
df9 = pd.read_sql(query_9, con=engine)
df9
```

```python
#Insurance Coverage Analysis - Calculates what percentage of total hospital bills are covered by insurance, helping assess financial dependency on insurance providers.
query_10 = "SELECT (SUM(`Health Insurance Amount`) / SUM(`Billing Amount`)) * 100 AS Insurance_Coverage_Percentage FROM patients_data;"
df10 = pd.read_sql(query_10, con=engine)
df10
```

```python
# Most Common Tests Ordered - Identifies the most frequently conducted tests, which helps optimize lab resources and manage test inventories.
query_11 = "SELECT Test, COUNT(*) AS Test_Count FROM patients_data GROUP BY Test ORDER BY Test_Count DESC LIMIT 5;"
df11 = pd.read_sql(query_11, con=engine)
df11
```

```python
#Fastest Discharges (Shortest Stay) 
query_12 = "SELECT Diagnosis, MIN(DATEDIFF(Discharge_Date, Admit_Date)) AS Min_Stay FROM patients_data GROUP BY Diagnosis ORDER BY Min_Stay ASC LIMIT 5;"
df12 = pd.read_sql(query_12, con=engine)
df12
```

```python
# Patient Satisfaction Analysis
query_13 = "SELECT Doctor, AVG(Feedback) AS Avg_Feedback FROM patients_data GROUP BY Doctor ORDER BY Avg_Feedback DESC;"
df13 = pd.read_sql(query_13, con=engine)
df13
```

```python
# Readmission Analysis
query_14 = "SELECT COUNT(DISTINCT Patient_ID) AS Total_Patients, COUNT(*) AS Total_Admissions, (COUNT(*) - COUNT(DISTINCT Patient_ID)) AS Readmissions FROM patients_data;"
df14 = pd.read_sql(query_14, con=engine)
df14
```

```python
#High-Risk Patients (Long Stays & High Bills)
query_15 = "SELECT Patient_ID, Diagnosis, DATEDIFF(Discharge_Date, Admit_Date) AS Length_Of_Stay, `Billing Amount` FROM patients_data WHERE DATEDIFF(Discharge_Date, Admit_Date) > (SELECT AVG(DATEDIFF(Discharge_Date, Admit_Date)) FROM patients_data) AND `Billing Amount` > (SELECT AVG(`Billing Amount`) FROM patients_data) ORDER BY `Billing Amount` DESC;"
df15 = pd.read_sql(query_15, con=engine)
df15
```

```python
query_16 = "SELECT Diagnosis, count(Diagnosis) AS no_of_cases from patients_data group by Diagnosis;"
df16 = pd.read_sql(query_16, con=engine)
df16
```
